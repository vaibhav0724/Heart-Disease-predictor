## Model selection

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.svm import SVC 
class1 = SVC(kernel = 'linear',random_state=0)
class1.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_p = class1.predict(X_test)
c = confusion_matrix(Y_test,y_p)
c
```

```python
from sklearn.naive_bayes import GaussianNB
v = GaussianNB()
v.fit(X_train,Y_train)
y_p1 = v.predict(X_test)
conf1 = confusion_matrix(Y_test,y_p1)
conf1
```

```python
v.score(X_test,Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,Y_train)
y_p2 = dtree.predict(X_test)
```

```python
conf2 = confusion_matrix(Y_test,y_p2)
conf2
```

```python
dtree.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfm = RandomForestClassifier()
rfm.fit(X_train, Y_train)
y_p3 = rfm.predict(X_test)
```

```python
conf3 = confusion_matrix(Y_test,y_p3)
conf3
```

```python
rfm.score(X_test,Y_test)
```

```python
print("Classification report is:\n\n ",classification_report(Y_test,y_p3))
```

## Best Among calculated Algorithm score I got best for Random Forest Classifier
